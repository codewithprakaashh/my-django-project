```python
for col in df.columns:
    if df[col].nunique() > 0:  # avoid errors for all-missing columns
        most_freq_value = df[col].mode().iloc[0]
        # get percentage of most frequent value
        most_freq_percentage = df[col].value_counts(normalize=True).iloc[0] * 100
        most_freq_count = df[col].value_counts().iloc[0]
        print(f"{col}: {most_freq_value} (Count: {most_freq_count})")
```
